```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sns
```

```python
dataFrame = pd.read_csv("d:/Programming Projects/PROJECTS_/Customer Churn Prediction/Data/Telco_Customer_Chern_Cleaned.csv")
```

```python
print("Shape: ", dataFrame.shape)
print("\nFirst 5 Rows:")
display(dataFrame.head())
```

```python
print("\nInfo:")
display(dataFrame.info())
```

```python
print("\nSummary Statistics:")
display(dataFrame.describe())
```

```python
print("\nChurn Value Counts:")
display(dataFrame['Churn_Yes'].value_counts())
```

```python
categorical_cols = dataFrame.select_dtypes(include='object').columns.tolist()
numerical_cols = dataFrame.select_dtypes(include=np.number).columns.tolist()
```

```python
for col in categorical_cols:
    plot.figure(figsize=(6, 4))
    sns.countplot(data=dataFrame, x=col, hue='Churn')
    plot.title(f"{col} vs Churn")
    plot.xticks(rotation=45)
    plot.tight_layout()
    plot.savefig(f"d:/Programming Projects/PROJECTS_/Customer Churn Prediction/Visuals/{col}_vs_Churn.png")
    plot.close()
```

```python
for col in numerical_cols:
    if col != 'Churn_Yes':  # avoid plotting target as a feature
        plot.figure(figsize=(6, 4))
        sns.histplot(data=dataFrame, x=col, hue='Churn_Yes', kde=True, element="step")
        plot.title(f"{col} Distribution by Churn")
        plot.tight_layout()
        plot.savefig(f"d:/Programming Projects/PROJECTS_/Customer Churn Prediction/Visuals/{col}_distribution_by_Churn.png")
        plot.close()
```

```python
plot.figure(figsize=(10, 8))
sns.heatmap(dataFrame[numerical_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plot.title("Correlation Heatmap (Numerical Features)")
plot.tight_layout()
plot.savefig("d:/Programming Projects/PROJECTS_/Customer Churn Prediction/Visuals/correlation_heatmap.png")
plot.close()
```

```python
if 'Churn' in dataFrame.columns and dataFrame['Churn'].dtype == 'object':
    churn_rate = dataFrame['Churn'].value_counts(normalize=True).get('Yes', 0) * 100
elif 'Churn_Yes' in dataFrame.columns:
    churn_rate = dataFrame['Churn_Yes'].mean() * 100
else:
    churn_rate = None
```

```python
findings = f"""
1. Dataset contains {dataFrame.shape[0]} rows and {dataFrame.shape[1]} columns.
2. Churn distribution is imbalanced with approx {churn_rate:.2f}% churn rate.
3. Tenure, MonthlyCharges, and Contract type show clear patterns with churn.
4. Customers with month-to-month contracts, no dependents, and low tenure have higher churn.
"""

with open("d:/Programming Projects/PROJECTS_/Customer Churn Prediction/Reports/EDA_Findings.txt", "w") as f:
    f.write(findings)

print("EDA complete. Churn rate calculated:", churn_rate)
```
